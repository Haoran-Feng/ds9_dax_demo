## Catalogs


### Data

This demo uses data for the Orion Nebula and data from the Chandra Source Catalog v1.1

[ROSETTE NEBULA / NGC 2244, OBS_ID 3750](http://cda.harvard.edu/srservices/ocatDetails.do?obsid=3750)

```bash
obsid_search_csc 3750 3750.tsv col='o.ks_prob_b,o.photflux_aper_b,o.cnts_aper_b'  clob+
download_chandra_obsid 3750 evt2
```

````bash
cd Catalog
ds9 acisf03750N003_evt2.fits -catalog import tsv 3750.tsv &
````

````bash
xpaset -p ds9 scale log
xpaset -p ds9 scale limits 0 100

import -window "SAOImage ds9" ds9_10_cat_01.png
display < ds9_10_cat_01.png 
````

````bash
# Grab catalog window

sleep 3
import  ds9_10_cat_02.png 
display < ds9_10_cat_02.png 
````

````bash
# 
# Click Plot button
#   X-Axis:  cnts_aper_b
#   Y-Axis:  photflux_aper_b
#
xpaset -p ds9 catalog plot 'cnts_aper_b ' , '{$photflux_aper_b}'  || echo
````

````bash
# Grab plot window

sleep 3
import -window "3750.tsv" ds9_10_cat_03.png 
display < ds9_10_cat_03.png 
````

````bash
xpaset -p ds9 catalog symbol load ds9.sym
````

````bash
# Symbol -> Advanced
sleep 5

import -window "Symbol Editor" ds9_10_cat_04.png 
display < ds9_10_cat_04.png 
````

````bash
xpaset -p ds9 zoom to 8
xpaset -p ds9 scale limits 0 100
xpaset -p ds9 pan to 4139 4123 physical
xpaset -p ds9 scale asinh

import -window "SAOImage ds9" ds9_10_cat_05.png 
display < ds9_10_cat_05.png 
````

````bash
xpaset -p ds9 quit
````
